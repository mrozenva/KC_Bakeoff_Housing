###### ANOVA of Property Prices using Top 5 Neighborhoods

```python
alpha = 0.05 
#ANOVA Test Setup
kc_top5 = df[df.zipcode.isin(kc_top5_price.index)]
formula = 'price~C(zipcode)'
lm_top5 = smf.ols(formula, kc_top5).fit()
anova_top5_table = sm.stats.anova_lm(lm_top5, typ=2)

if anova_top5_table["PR(>F)"][0] < alpha:
    print("Top 5 ranked zipcode have a statistically significant impact on average property value")
    print("Zipcdoe ANOVA F-statisic Probability: ", anova_top5_table["PR(>F)"][0])
```

#### Lets take a look at all the zipcodes not just the top 5

##### ANOVA Test

#### Rationale: 
Comparing means of several independent samples
#### Ho:
There is no difference in the average price of houses in different zip codes.
#### Ha: 
Houses in different zip codes have different average house prices

```python
# Create dummy variables for categorical zip code column
zip_dummies = pd.get_dummies(df['zipcode'].astype(str), dtype=int, drop_first=True)
df.drop(columns=['zipcode'],inplace=True)
new_cols = 'zip'+zip_dummies.columns
zip_dummies.columns = new_cols
zip_dummies.head()
```

```python
# Concatenate zipcode dummy variables with with main dataframe
X_dummies = pd.concat([df['price'], zip_dummies], axis=1)
formula = "price ~ C("+ ') + C('.join(list(zip_dummies.columns))+')'
model = ols(formula, X_dummies).fit()
table = sm.stats.anova_lm(model, typ=2)
table.sort_values('PR(>F)').head(10)
```

###### Very low P-values for all zip codes indicate that we can reject the null hypothesis in favor of the alternative, that houses in different zip codes have different prices on average.

```python
#--------------------------Property Feature Calculation---------------------------------------#
basement = df[(df["sqft_basement"] > 0)]
basement_mean = basement.price.mean()
no_basement = df[(df["sqft_basement"] == 0)]
no_basement_mean = no_basement.price.mean()

#mean values to plot
renovated = df[(df["yr_renovated"] > 0)]
renovated_mean = renovated.price.mean()
not_renovated = df[(df["yr_renovated"] == 0)]
not_renovated_mean = not_renovated.price.mean()

#prepare plot labels
label_basement = ["Basement", "No basement"]
values_basement = [basement_mean, no_basement_mean]
label_renovation = ["Renovated", "No Renovation"]
values_renovation = [renovated_mean, not_renovated_mean]

#--------------------------Bar Plots--------------------------------------#


sns.barplot(ax=ax[0], x=label_basement, y=values_basement, palette="Blues_r")
sns.barplot(ax=ax[1], x=label_renovation, y=values_renovation, palette="Blues_r")
ax[0].hlines(kc_mean_price, -.5 ,1.5, colors="coral", label="Average Price") #plot average price horizontal line
ax[1].hlines(kc_mean_price, -.5 ,1.5, colors="coral", label="Average Price") #plot average price horizontal line

#prettify graph
ax[0].set_ylabel("Average Prices ($)", size=12)
ax[0].set_title("Average Property Value", size=14)
ax[0].set_ylim(0,1.1*renovated_mean)
ax[0].legend()

ax[1].set_ylabel("Average Prices ($)", size=12)
ax[1].set_title("Average Property Value", size=14)
ax[1].set_ylim(0,1.1*renovated_mean)
ax[1].legend()

plt.suptitle("Affect of Basement and Renovation on Property Value", size=16, y=1.02)

#uncomment below to export image
# plt.savefig("images/basement_renovation_value.png",bbox_inches = "tight")
plt.show();
```

#### 2-Sample Welch's T-Test For Basement (significance level: 0.05)

##### Ho: 
There is no significant difference in the average price of houses with and without basements.
##### Ha:
Houses with basements sell for higher prices on average.

```python
alpha = 0.05
basement_p_val = stats.ttest_ind(basement.price, no_basement.price, equal_var=False)[1]
print("Basement vs No Basement T-test P Value: ", basement_p_val)
if basement_p_val < alpha:
    print("The P value is less than alpha, reject null-hypothesis")

print("\n") #white space for formatting output
```

#### 2-Sample Welch's T-Test For Renovations (significance level: 0.05)

###### Ho:
There is no difference in the average price of houses with and without renovations as well as how long ago it was renovated.

###### Ha:
Houses with newer renovations sell for higher prices on average.

```python
renovation_p_val = stats.ttest_ind(renovated.price, not_renovated.price, equal_var=False)[1]
print("Renovated vs Not Renovated T-test P Value: ", renovation_p_val)
if renovation_p_val < alpha:
    print("The P value is less than alpha, reject null-hypothesis")
```

### Both basement and renovations add value to the property, as both null hypothesis's were accepted.

### Lets get an R-squared for every interaction

```python
# Generate combinations of features
X = df.iloc[:,1:]
interactions = list(combinations(X.columns, 2))
interaction_dict = {}
y = df['price']
for interaction in interactions:
    X_copy = X.copy()
    X_copy['interact'] = X_copy[interaction[0]] * X_copy[interaction[1]]  
    model = LinearRegression() #run model with each possible interaction
    model.fit(X_copy, y)
    interaction_dict[model.score(X_copy, y)] = interaction # add R-squared for each interaction to a dictionary
sorted(interaction_dict.items(), reverse = True)[:10]
```

#### Add 54 best interactions to our model

```python
interaction_df = X.copy()

top_interactions = sorted(interaction_dict.keys(), reverse = True)[:54]

for interaction in top_interactions:
    feature1 = interaction_dict[interaction][0]
    feature2 = interaction_dict[interaction][1]
    interaction_df[feature1+'_X_'+feature2] = interaction_df[feature1] * interaction_df[feature2]
    new_features[feature1+'_X_'+feature2] = interaction_df[feature1] * interaction_df[feature2]
    model = LinearRegression() # run model with each added interaction term
    model.fit(interaction_df, y)
    r_squared = model.score(interaction_df, y)
    print('R-Squared is',r_squared,'after adding:  ',feature1+'_X_'+feature2)
```
